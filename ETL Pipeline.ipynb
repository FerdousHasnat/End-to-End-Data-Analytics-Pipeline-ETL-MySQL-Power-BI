```python
import pandas as pd

df = pd.read_csv("C:\\Users\\3070\\Desktop\\ETL PowerBI\\Superstore.csv", encoding='ISO-8859-1')

df.head()

```

```python
# Rename columns clearly
df.columns = df.columns.str.replace(' ', '_').str.replace('-', '_')

# Confirm changes
print(df.columns.tolist())
```

```python
# Convert date columns clearly
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

# Convert Postal_Code clearly into categorical (string type)
df['Postal_Code'] = df['Postal_Code'].astype(str)

# Verify data types explicitly
print(df.dtypes)
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Duplicates before removal: {duplicates}")

# Remove duplicates (if any)
df.drop_duplicates(inplace=True)

# Check again
print(f"Data shape after duplicate removal: {df.shape}")
```

```python
from sqlalchemy import create_engine

# MySQL connection details
username = 'root'
password = 'Nayeem32!'
database_name = 'sales_insights'

# Create database connection string
db_connection_str = f'mysql+pymysql://{username}:{password}@localhost/{database_name}'

# Create SQLAlchemy engine
engine = create_engine(db_connection_str)

# Check connection
try:
    connection = engine.connect()
    print("Successfully connected to MySQL!")
    connection.close()
except Exception as e:
    print(f"Error: {e}")
```

```python
# Loading data clearly into MySQL
df.to_sql('superstore_sales', engine, if_exists='replace', index=False)

print("Data loaded successfully into MySQL!")
```
